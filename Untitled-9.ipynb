```python
# Text cleaning function
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Sample and clean data
df_sampled = df.sample(n=20000, random_state=42)
cleaned_reviews = [clean_text(review) for review in df_sampled['description'].tolist()]
print(f"Cleaned {len(cleaned_reviews)} reviews")
```

### 1.1 n-gram --- many to one

```python
# N-gram Model Definition
def build_ngram_model(tokens, n=3):
    """Build n-gram probability model"""
    model = defaultdict(Counter)
    for i in range(len(tokens) - n + 1):
        context = tuple(tokens[i : i + n - 1])
        target = tokens[i + n - 1]
        model[context][target] += 1
    
    prob_model = {}
    for context, counts in model.items():
        total = sum(counts.values())
        prob_model[context] = {word: count / total for word, count in counts.items()}
    return prob_model

def calculate_ngram_ppl(model, test_tokens, vocab_size, n=3):
    """Calculate perplexity on test set"""
    log_prob_sum = 0
    count = 0
    for i in range(len(test_tokens) - n + 1):
        context = tuple(test_tokens[i : i + n - 1])
        target = test_tokens[i + n - 1]
        
        if context in model and target in model[context]:
            prob = model[context][target]
        else:
            prob = 1 / vocab_size 
            
        log_prob_sum += math.log2(prob)
        count += 1
    
    avg_log_prob = log_prob_sum / count
    return 2 ** (-avg_log_prob)

def evaluate_ngram_accuracy(model, test_tokens, n=3):
    """Evaluate accuracy on test set"""
    correct = 0
    total = 0
    
    for i in range(len(test_tokens) - n + 1):
        context = tuple(test_tokens[i : i + n - 1])
        actual_next = test_tokens[i + n - 1]
        
        if context in model:
            predicted_next = max(model[context].items(), key=lambda x: x[1])[0]
            if predicted_next == actual_next:
                correct += 1
        total += 1
    
    accuracy = correct / total if total > 0 else 0
    return accuracy, correct, total

def generate_text_ngram(model, start_context, gen_len=20):
    """Generate text using n-gram model"""
    generated_ids = list(start_context)
    curr_context = start_context
    
    start_inf = time.time()
    for _ in range(gen_len):
        if curr_context in model:
            choices = list(model[curr_context].keys())
            weights = list(model[curr_context].values())
            next_id = random.choices(choices, weights=weights)[0]
            generated_ids.append(next_id)
            curr_context = tuple(generated_ids[-2:])
        else:
            break
    inf_time = time.time() - start_inf
    
    text = [idx2word.get(idx, "<UNK>") for idx in generated_ids]
    return " ".join(text), inf_time

print("✅ N-gram functions defined")
```

### 1.2 RNN - many to many

```python
# Dataset for Many-to-Many Training
class LanguageModelDataset(Dataset):
    def __init__(self, data, seq_len):
        self.data = data
        self.seq_len = seq_len
        
    def __len__(self):
        return len(self.data) - self.seq_len
    
    def __getitem__(self, idx):
        # Input: [w1, w2, ..., w20]
        # Target: [w2, w3, ..., w21] - shifted sequence
        x = torch.LongTensor(self.data[idx : idx + self.seq_len])
        y = torch.LongTensor(self.data[idx + 1 : idx + self.seq_len + 1])
        return x, y

print("Dataset class defined (Many-to-Many targets)")
```

```python
# Text Generation Function (shared by RNN and LSTM)
def generate_text_neural(model, start_tokens, gen_len=20, temperature=1.0):
    """Generate text using neural model"""
    model.eval()
    generated = list(start_tokens)
    
    start_inf = time.time()
    with torch.no_grad():
        for _ in range(gen_len):
            x = torch.LongTensor([generated[-SEQ_LEN:]]).to(device)
            output = model(x)  # (1, seq_len, vocab_size)
            logits = output[0, -1, :] / temperature  # Last position
            probs = torch.softmax(logits, dim=0)
            next_token = torch.multinomial(probs, 1).item()
            generated.append(next_token)
    
    inf_time = time.time() - start_inf
    text = [idx2word.get(idx, "<UNK>") for idx in generated]
    return " ".join(text), inf_time

print("✅ generate_text_neural function defined (shared by RNN and LSTM)")
```

```python
# Training and Evaluation Functions
def train_epoch(model, dataloader, criterion, optimizer, device):
    """Train for one epoch with Many-to-Many loss"""
    model.train()
    total_loss = 0
    for x, y in dataloader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        
        output = model(x)  # (batch, seq_len, vocab_size)
        
        # Reshape for loss calculation: flatten batch and sequence dimensions
        loss = criterion(output.reshape(-1, output.size(-1)), y.reshape(-1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    return total_loss / len(dataloader)

def evaluate(model, dataloader, criterion, device):
    """Evaluate model - only check last position prediction for fair comparison"""
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            output = model(x)  # (batch, seq_len, vocab_size)
            
            # Only evaluate last position (for fair comparison with N-gram)
            output_last = output[:, -1, :]  # (batch, vocab_size)
            y_last = y[:, -1]  # (batch,)
            
            loss = criterion(output_last, y_last)
            total_loss += loss.item()
            
            pred = output_last.argmax(dim=1)
            correct += (pred == y_last).sum().item()
            total += y_last.size(0)
    
    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total
    perplexity = math.exp(avg_loss)
    
    return avg_loss, accuracy, perplexity

print("✅ Training/evaluation functions defined")
```

```python
# RNN Model Definition
class RNNLanguageModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_layers):
        super(RNNLanguageModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
        
    def forward(self, x):
        # x: (batch, seq_len)
        embedded = self.embedding(x)  # (batch, seq_len, embedding_dim)
        output, _ = self.rnn(embedded)  # (batch, seq_len, hidden_dim)
        output = self.fc(output)  # (batch, seq_len, vocab_size) - Many-to-Many
        return output

print("✅ RNN model class defined (Many-to-Many architecture)")
```

```python
# RNN Hyperparameters
SEQ_LEN = 20
BATCH_SIZE = 128
EMBEDDING_DIM = 128
HIDDEN_DIM = 256
NUM_LAYERS = 2
LEARNING_RATE = 0.001
NUM_EPOCHS = 5

print("RNN Configuration:")
print(f"  Architecture: Many-to-Many (20 predictions per sequence)")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Sequence Length: {SEQ_LEN}")
print(f"  Embedding Dim: {EMBEDDING_DIM}")
print(f"  Hidden Dim: {HIDDEN_DIM}")
print(f"  Num Layers: {NUM_LAYERS}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Epochs: {NUM_EPOCHS}")
```

```python
# Create Data Loaders
train_dataset = LanguageModelDataset(train_data, SEQ_LEN)
test_dataset = LanguageModelDataset(test_data, SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"✅ Data loaders created:")
print(f"   Train batches: {len(train_loader)}")
print(f"   Test batches: {len(test_loader)}")
```

### 1.4 Transformer model （fixed lr & many to many）

```python
# Transformer Components
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

class TransformerLanguageModel(nn.Module):
    def __init__(self, vocab_size, d_model, nhead, num_layers, dim_feedforward):
        super(TransformerLanguageModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.pos_encoder = PositionalEncoding(d_model)
        encoder_layer = nn.TransformerEncoderLayer(d_model, nhead, dim_feedforward, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers)
        self.fc = nn.Linear(d_model, vocab_size)
        self.d_model = d_model
        
    def forward(self, x):
        # x: (batch, seq_len)
        embedded = self.embedding(x) * math.sqrt(self.d_model)  # (batch, seq_len, d_model)
        embedded = self.pos_encoder(embedded)
        
        # Create causal mask
        seq_len = x.size(1)
        mask = torch.triu(torch.ones(seq_len, seq_len), diagonal=1).bool().to(x.device)
        
        output = self.transformer(embedded, mask=mask)  # (batch, seq_len, d_model)
        output = self.fc(output)  # (batch, seq_len, vocab_size) - Many-to-Many
        return output

print("✅ Transformer model class defined (Many-to-Many architecture)")
```

```python
# Transformer Hyperparameters
D_MODEL = 128
NHEAD = 4
NUM_TRANSFORMER_LAYERS = 2
DIM_FEEDFORWARD = 512

print("Transformer Configuration:")
print(f"  Architecture: Many-to-Many (20 predictions per sequence)")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  d_model: {D_MODEL}")
print(f"  nhead: {NHEAD}")
print(f"  Num Layers: {NUM_TRANSFORMER_LAYERS}")
print(f"  Feedforward Dim: {DIM_FEEDFORWARD}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Epochs: {NUM_EPOCHS}")
```

#### 1.6 Transformer model （warmup & many to many）

```python
# WarmUp + CosineAnnealing Scheduler
class WarmUpCosineScheduler:
    def __init__(self, optimizer, warmup_epochs, total_epochs, base_lr):
        self.optimizer = optimizer
        self.warmup_epochs = warmup_epochs
        self.total_epochs = total_epochs
        self.base_lr = base_lr
        
    def step(self, epoch):
        if epoch < self.warmup_epochs:
            # WarmUp phase: linear increase from 0 to base_lr
            lr = self.base_lr * (epoch + 1) / self.warmup_epochs
        else:
            # CosineAnnealing phase
            progress = (epoch - self.warmup_epochs) / (self.total_epochs - self.warmup_epochs)
            lr = self.base_lr * 0.5 * (1 + math.cos(math.pi * progress))
        
        for param_group in self.optimizer.param_groups:
            param_group['lr'] = lr
        
        return lr

print("✅ WarmUp + CosineAnnealing scheduler defined")
```

```python
# GloVe-style Embedding Implementation
class GloVeModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(GloVeModel, self).__init__()
        self.target_embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.context_embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.target_biases = nn.Embedding(vocab_size, 1)
        self.context_biases = nn.Embedding(vocab_size, 1)
        
        # Initialize
        nn.init.normal_(self.target_embeddings.weight, mean=0, std=0.01)
        nn.init.normal_(self.context_embeddings.weight, mean=0, std=0.01)
        nn.init.zeros_(self.target_biases.weight)
        nn.init.zeros_(self.context_biases.weight)
    
    def forward(self, target_idx, context_idx, cooccur, weight):
        # target_idx, context_idx: (batch_size,)
        # cooccur: (batch_size,) - log of co-occurrence count
        # weight: (batch_size,) - weighting function
        
        target_embed = self.target_embeddings(target_idx)  # (batch_size, embed_dim)
        context_embed = self.context_embeddings(context_idx)  # (batch_size, embed_dim)
        target_bias = self.target_biases(target_idx).squeeze()  # (batch_size,)
        context_bias = self.context_biases(context_idx).squeeze()  # (batch_size,)
        
        # Predicted score
        pred = torch.sum(target_embed * context_embed, dim=1) + target_bias + context_bias
        
        # Weighted squared error
        loss = weight * torch.pow(pred - cooccur, 2)
        return torch.mean(loss)

def build_cooccurrence_matrix(corpus_indices, vocab_size, window_size=5):
    """Build co-occurrence matrix"""
    cooccur_matrix = defaultdict(Counter)
    
    for i in range(len(corpus_indices)):
        target = corpus_indices[i]
        context_start = max(0, i - window_size)
        context_end = min(len(corpus_indices), i + window_size + 1)
        
        for j in range(context_start, context_end):
            if i != j:
                context = corpus_indices[j]
                distance = abs(i - j)
                cooccur_matrix[target][context] += 1.0 / distance
    
    return cooccur_matrix

def weighting_function(x, x_max=100, alpha=0.75):
    """GloVe weighting function"""
    return np.minimum(1.0, np.power(x / x_max, alpha))

print("✅ GloVe model defined")
```

```python
# Train GloVe Embeddings
print("=" * 60)
print("Training GloVe-style Embeddings")
print("=" * 60)

# Build co-occurrence matrix
print("Building co-occurrence matrix...")
cooccur_start = time.time()
cooccur_matrix = build_cooccurrence_matrix(train_data, vocab_size, window_size=5)
cooccur_time = time.time() - cooccur_start
print(f"✅ Co-occurrence matrix built in {cooccur_time:.2f}s")

# Prepare training data
train_pairs = []
for target, contexts in cooccur_matrix.items():
    for context, count in contexts.items():
        if count > 0:
            train_pairs.append((target, context, np.log(count), weighting_function(count)))

print(f"   Total pairs: {len(train_pairs):,}")

# Convert to tensors
targets = torch.LongTensor([p[0] for p in train_pairs])
contexts = torch.LongTensor([p[1] for p in train_pairs])
cooccurs = torch.FloatTensor([p[2] for p in train_pairs])
weights = torch.FloatTensor([p[3] for p in train_pairs])

# Create dataset
class GloVeDataset(Dataset):
    def __init__(self, targets, contexts, cooccurs, weights):
        self.targets = targets
        self.contexts = contexts
        self.cooccurs = cooccurs
        self.weights = weights
    
    def __len__(self):
        return len(self.targets)
    
    def __getitem__(self, idx):
        return self.targets[idx], self.contexts[idx], self.cooccurs[idx], self.weights[idx]

glove_dataset = GloVeDataset(targets, contexts, cooccurs, weights)
glove_loader = DataLoader(glove_dataset, batch_size=1024, shuffle=True)

# Initialize model
GLOVE_EMBEDDING_DIM = 128
GLOVE_EPOCHS = 5
GLOVE_LR = 0.05

glove_model = GloVeModel(vocab_size, GLOVE_EMBEDDING_DIM).to(device)
glove_optimizer = torch.optim.Adagrad(glove_model.parameters(), lr=GLOVE_LR)

# Training loop
start_time = time.time()
for epoch in range(GLOVE_EPOCHS):
    epoch_start = time.time()
    total_loss = 0
    glove_model.train()
    
    for batch_idx, (target, context, cooccur, weight) in enumerate(glove_loader):
        target = target.to(device)
        context = context.to(device)
        cooccur = cooccur.to(device)
        weight = weight.to(device)
        
        glove_optimizer.zero_grad()
        loss = glove_model(target, context, cooccur, weight)
        loss.backward()
        glove_optimizer.step()
        
        total_loss += loss.item()
        
        if (batch_idx + 1) % 1000 == 0:
            print(f"  Batch {batch_idx+1}/{len(glove_loader)}, Loss: {loss.item():.4f}")
    
    epoch_time = time.time() - epoch_start
    avg_loss = total_loss / len(glove_loader)
    print(f"Epoch {epoch+1}/{GLOVE_EPOCHS}")
    print(f"  Avg Loss: {avg_loss:.4f}")
    print(f"  Time: {epoch_time:.2f}s")

total_time = time.time() - start_time
print(f"\n✅ GloVe training completed in {total_time:.2f}s")

# Extract trained embeddings (average target and context embeddings)
glove_embeddings = (
    glove_model.target_embeddings.weight.data.cpu().numpy() + 
    glove_model.context_embeddings.weight.data.cpu().numpy()
) / 2
print(f"   Embedding shape: {glove_embeddings.shape}")
```

```python
# LSTM with Configurable Embeddings
class LSTMWithEmbedding(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_layers, 
                 pretrained_embeddings=None, freeze_embeddings=False):
        super(LSTMWithEmbedding, self).__init__()
        
        # Embedding layer
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        
        if pretrained_embeddings is not None:
            # Load pretrained embeddings
            self.embedding.weight.data.copy_(torch.from_numpy(pretrained_embeddings))
            print(f"   Loaded pretrained embeddings: {pretrained_embeddings.shape}")
        
        if freeze_embeddings:
            # Freeze embedding weights
            self.embedding.weight.requires_grad = False
            print(f"   Embeddings frozen (not trainable)")
        else:
            print(f"   Embeddings trainable")
        
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
        
    def forward(self, x):
        embedded = self.embedding(x)
        output, _ = self.lstm(embedded)
        output = self.fc(output)
        return output

print("✅ LSTM with configurable embeddings defined")
```

```python
# Training function with loss tracking for convergence analysis
def train_with_tracking(model, train_loader, criterion, optimizer, device, num_epochs):
    """Train model and track loss for each epoch"""
    model.train()
    epoch_losses = []
    epoch_times = []
    
    for epoch in range(num_epochs):
        epoch_start = time.time()
        total_loss = 0
        
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            
            output = model(x)
            loss = criterion(output.reshape(-1, output.size(-1)), y.reshape(-1))
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        epoch_time = time.time() - epoch_start
        avg_loss = total_loss / len(train_loader)
        
        epoch_losses.append(avg_loss)
        epoch_times.append(epoch_time)
        
        print(f"Epoch {epoch+1}/{num_epochs}")
        print(f"  Train Loss: {avg_loss:.4f}")
        print(f"  Time: {epoch_time:.2f}s")
    
    return epoch_losses, epoch_times

print("✅ Training function with tracking defined")
```

```python
from transformers import AutoTokenizer, AutoModel
import numpy as np

print("=" * 60)
print("Preparing Pretrained Embeddings (Public Source: BERT-base-uncased)")
print("=" * 60)

# Step 1:
bert_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
bert_model_hf = AutoModel.from_pretrained("bert-base-uncased")
bert_model_hf.eval()
BERT_EMBEDDING_DIM = bert_model_hf.config.hidden_size  # 768

# Step 2:
bert_embeddings_matrix = bert_model_hf.embeddings.word_embeddings.weight.data.cpu().numpy()
bert_vocab = bert_tokenizer.get_vocab()

# Step 3:
embedding_matrix = np.random.normal(0, 0.01, (vocab_size, BERT_EMBEDDING_DIM))
found_words = 0
for word, idx in word2idx.items():
    bert_idx = bert_vocab.get(word, None)
    if bert_idx is not None:
        embedding_matrix[idx] = bert_embeddings_matrix[bert_idx]
        found_words += 1

print(f"✅ Embedding matrix ready: shape={embedding_matrix.shape}")
print(f"   Matched {found_words}/{vocab_size} words from BERT vocab")
print(f"   These embeddings will be frozen (not updated during training)")

# Step 4:
print("\n" + "=" * 60)
print("Training Variant 4: Fixed Public Pretrained Embeddings (BERT)")
print("=" * 60)

model_pretrained = LSTMWithEmbedding(
    vocab_size,
    BERT_EMBEDDING_DIM,
    ABLATION_HIDDEN_DIM,
    ABLATION_NUM_LAYERS,
    pretrained_embeddings=embedding_matrix,
    freeze_embeddings=True
).to(device)

optimizer_pretrained = torch.optim.Adam(model_pretrained.parameters(), lr=ABLATION_LR)

total_params = sum(p.numel() for p in model_pretrained.parameters())
trainable_params = sum(p.numel() for p in model_pretrained.parameters() if p.requires_grad)
print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,} (embeddings frozen)")

# Step 5:
start_time = time.time()
losses_pretrained, times_pretrained = train_with_tracking(
    model_pretrained, train_loader, criterion, optimizer_pretrained, device, ABLATION_EPOCHS
)
train_time_pretrained = time.time() - start_time

# Step 6:
test_loss_pretrained, test_acc_pretrained, test_ppl_pretrained = evaluate(
    model_pretrained, test_loader, criterion, device
)

print(f"\n✅ Variant 4 (BERT) completed in {train_time_pretrained:.2f}s")
print("📊 Test Results:")
print(f"   Loss: {test_loss_pretrained:.4f}")
print(f"   Accuracy: {test_acc_pretrained:.4f}")
print(f"   Perplexity: {test_ppl_pretrained:.2f}")
```

This ablation study systematically evaluates **five** embedding strategies for neural language modeling on a wine review dataset (20,000 samples, vocabulary size 7,529). **Key Distinction**: The **only** differences between variants are:
1. **Embedding Initialization**: V1 (random) vs V2-V5 (pretrained)

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

ds_tokenizer = word2idx
ds_seq_len = SEQ_LEN
ds_device = device

@torch.no_grad()
def ds_extract_lstm_reps(reviews, tokenizer, seq_len, model, device):
    """Extract LSTM last hidden state as sentence representation"""
    model.eval()
    reps = []
    for review in reviews:
        tokens = [tokenizer.get(w, tokenizer['<UNK>']) for w in review.split()]
        tokens = tokens[:seq_len] if len(tokens) >= seq_len else tokens + [tokenizer['<PAD>']] * (seq_len - len(tokens))
        t = torch.tensor(tokens, dtype=torch.long).unsqueeze(0).to(device)
        embedded = model.embedding(t)
        _, (h_n, _) = model.lstm(embedded)
        reps.append(h_n[-1, 0, :].cpu())
    return torch.stack(reps)

@torch.no_grad()
def ds_extract_rnn_reps(reviews, tokenizer, seq_len, model, device):
    """Extract RNN last hidden state as sentence representation"""
    model.eval()
    reps = []
    for review in reviews:
        tokens = [tokenizer.get(w, tokenizer['<UNK>']) for w in review.split()]
        tokens = tokens[:seq_len] if len(tokens) >= seq_len else tokens + [tokenizer['<PAD>']] * (seq_len - len(tokens))
        t = torch.tensor(tokens, dtype=torch.long).unsqueeze(0).to(device)
        embedded = model.embedding(t)
        _, h_n = model.rnn(embedded)
        reps.append(h_n[-1, 0, :].cpu())
    return torch.stack(reps)

@torch.no_grad()
def ds_extract_transformer_reps(reviews, tokenizer, seq_len, model, device):
    """Extract Transformer representations: full encoder forward pass (with pos_encoder),
    then apply mean pooling over all token outputs. Model uses batch_first=True.
    """
    model.eval()
    reps = []
    for review in reviews:
        tokens = [tokenizer.get(w, tokenizer['<UNK>']) for w in review.split()]
        tokens = tokens[:seq_len] if len(tokens) >= seq_len else tokens + [tokenizer['<PAD>']] * (seq_len - len(tokens))
        t = torch.tensor(tokens, dtype=torch.long).unsqueeze(0).to(device)
        # Embedding + positional encoding (batch_first=True, shape: (1, seq_len, d_model))
        emb = model.embedding(t) * math.sqrt(model.d_model)
        emb = model.pos_encoder(emb)
        # Pass through transformer encoder
        enc_out = model.transformer(emb)        # (1, seq_len, d_model)
        rep = enc_out.mean(dim=1).squeeze(0).cpu()  # mean pooling → (d_model,)
        reps.append(rep)
    return torch.stack(reps)

print("✅ Representation extraction functions defined")
```

```python
class DSClassifier(nn.Module):
    def __init__(self, input_dim, num_classes=2):
        super().__init__()
        self.fc = nn.Linear(input_dim, num_classes)
    def forward(self, x):
        return self.fc(x)

def ds_train_and_eval(train_loader, test_loader, input_dim, epochs=5, lr=1e-3, label=''):
    """Train downstream classifier and return test accuracy"""
    clf = DSClassifier(input_dim).to(ds_device)
    ds_criterion = nn.CrossEntropyLoss()
    ds_optimizer = optim.Adam(clf.parameters(), lr=lr)

    for epoch in range(epochs):
        clf.train()
        total_loss, correct, total = 0, 0, 0
        for reps, lbls in train_loader:
            reps, lbls = reps.to(ds_device), lbls.to(ds_device)
            out = clf(reps)
            loss = ds_criterion(out, lbls)
            ds_optimizer.zero_grad()
            loss.backward()
            ds_optimizer.step()
            total_loss += loss.item() * reps.size(0)
            correct += (out.argmax(1) == lbls).sum().item()
            total += reps.size(0)
        print(f"[{label}] Epoch {epoch+1}/{epochs}  Loss={total_loss/total:.4f}  Train Acc={correct/total:.4f}")

    clf.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for reps, lbls in test_loader:
            reps, lbls = reps.to(ds_device), lbls.to(ds_device)
            correct += (clf(reps).argmax(1) == lbls).sum().item()
            total += reps.size(0)
    test_acc = correct / total
    print(f"[{label}] Test Accuracy: {test_acc:.4f}\n")
    return test_acc

print("✅ Downstream classifier defined")
```
